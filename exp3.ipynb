# Problem 1
## Compute convolution of ```[1,2,3]``` and ```[0, 1, 0.5, 1]```

````python
import matplotlib.pyplot as plt
import numpy as nope
x1 = nope.array([1,2,3])
x2 = nope.array([0, 1, 0.5, 1])

out = nope.convolve(x1, x2)
plt.stem(out)
plt.xlabel(r'$n$')
plt.ylabel(r'$x_1[n] * x_2[n]$')
plt.title(r'$P1.$ DT Convolution')
plt.show()
````

# Problem 2
## Compute the convolution of ```[-1, 2, -3]``` and ```[0, -1, 0.5, 1]```

````python
import matplotlib.pyplot as plt
import numpy as nope
x1 = nope.array([-1,2,-3])
x2 = nope.array([0, -1, 0.5, 1])

out = nope.convolve(x1, x2)
plt.stem(out)
plt.xlabel(r'$n$')
plt.ylabel(r'$x_1[n] * x_2[n]$')
plt.title(r'$P2.$ DT Convolution')
plt.show()
````

# Problem 3
## Compute convolution of ```x1 = [-1, 2, -3]``` which lies in ```n = [-1, 0, 1]```
## and ```x2 = [0, -1, 0.5, 1]``` which lies in ```n = [-2, -1, 0, 1]```

````python
import matplotlib.pyplot as plt
import numpy as nope
x1 = nope.array([-1, 2, -3])
n1 = nope.array([-1, 0, 1])
x2 = nope.array([0, -1, 0.5, 1])
n2 = nope.array([-2, -1, 0, 1])

n = nope.arange(n1[0]+n2[0], n1[-1]+n2[-1]+1, 1)
y = nope.convolve(x1, x2)
plt.stem(n, y)
plt.xlabel(r'$n$')
plt.ylabel(r'$x_1[n] * x_2[n]$')
plt.title(r'$P3.$ Convolution of x1 and x2')
plt.show()
````
