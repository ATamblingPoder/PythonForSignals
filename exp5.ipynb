````python
import matplotlib.pyplot as plt
import numpy as nope
from numpy import pi as pi
````

````python
def plotter(start, end, incr, func, label_p=None, title_p=None, x_p=None, y_p=None, plot=1):
  t = nope.arange(start, end, incr)
  func = nope.vectorize(func)
  x1 = func(t)
  plt.figure(figsize=(8,2))
  if plot:
    plt.plot(t, x1, label=label_p)
  else:
    plt.stem(t, x1, label=label_p)
  plt.title(title_p)
  plt.xlabel(x_p)
  plt.ylabel(y_p)
  plt.legend()
  plt.show()


def plotter2(t_in, func, ylab):
    plt.figure(figsize=(6,1))
    plt.plot(t_in, func)
    plt.xlabel(r"$t$")
    plt.ylabel(ylab)
    plt.show()
````

# In-Lab Exercises
## 1. ```x1(t) = cos(6πt)``` for ```0 ≤ t ≤ 1```

````python
q1 = lambda i: nope.cos(6*pi*i)
plotter(0, 1, 0.01, q1, r"$cos(6 \pi t)$", r"Cosine Curve", r"$t$", r"$cos(6 \pi t)$")
````

## 2. ```x2(t) = 3cos(3πt + π/3)```

````python
q2 = lambda i: 3*nope.cos(3*pi*i + (pi/3))
plotter(0, 3, 0.01, q2, r"$3cos(3 \pi t + \frac{\pi}{3})$", r"Cosine Curve", r"$t$", r"$3cos(3 \pi t + \frac{\pi}{3})$")
````

## 3. ```x3(t) = 3e^0.4t``` and ```x4(t) = 2e^-0.9t``` in time interval ```-2 ≤ t ≤ 6```

````python
t = nope.arange(-2,6,0.01)
x3 = 3*nope.exp(0.4*t)
x4 = 2*nope.exp(-.9*t)
plt.figure(figsize = (8,2))
plt.plot(t,x3,label = r'$x_3(t)= 3e^{0.4t}$')
plt.plot(t,x4, label = r'$x_4(t) = 2e^{-0.9t}$')
plt.ylabel('Exponential Signals')
plt.title('Exponential Signals')
plt.xlabel(r'$t$')
plt.legend()
plt.show()
````

## 4. Code for ```g[n]```

````python
q4 = lambda i: 5*(i<=0) + (5-3*i)*(i>0)*(i<=4) + (-23 + i*i)*(i>4)*(i<=8) + 41*(i>8)
plotter(-12, 13, 1, q4, "$g[n]$", r"$g[n]$", r"$n$", r"$g[n]$", 0)
````

## 5. ```x1(t) + x2(t)``` for ```-1 ≤ t ≤ 2```

````python
q5 = lambda i: nope.cos(6*pi*i) + 3*nope.cos(3*pi*i + (pi/3))
plotter(-1, 2, 0.01, q5, r"$x_1(t) + x_2(t)$", "Addition", r"$t$", r"$x_1(t) + x_2(t)$")
````

## 6. Find even and odd components of signal ```x_3(t)``` as defined in Problem 2

````python
t = nope.arange(-2, 2, 0.2)
q3 = lambda i: 3*nope.exp(0.4*i)
q3 = nope.vectorize(q3)
q3_f = q3(t)
q3_e = (q3(t) + q3(-t))/2
q3_o = (q3(t) - q3(-t))/2
plt.plot(t, q3_e, label=r"Even Part of $x(t)$")
plt.plot(t, q3_o, label=r"Odd Part of $x(t)$")
plt.plot(t, q3_e+q3_o, 'r*-', label=r"Reconstructed $x(t)$")
plt.xlabel(r"$t$")
plt.legend()
plt.show()
````

## 7. Generate ```x(t) = t*e^(-t)``` for ```-1 <= t <= 1```, plot the following signals
1. ```x(t)```
2. ```x(-t)```
3. ```x(t/5)```
4. ```x(5t)```
5. ```x(t-3)```
6. ```x(1-2t)```

````python
t = nope.arange(-1, 1, 0.01)
xt = lambda i: i * nope.exp(-i)
x_1 = xt(t)
x_2 = xt(-t)
x_3 = xt(t/5)
x_4 = xt(5*t)
x_5 = xt(t - 3)
x_6 = xt(1 - 2*t)
plotter2(t, x_1, r"$x(t)$")
plotter2(t, x_2, r"$x(-t)$")
plotter2(t, x_3, r"$x(\frac{t}{5})$")
plotter2(t, x_4, r"$x(5t)$")
plotter2(t, x_5, r"$x(t - 3)$")
plotter2(t, x_6, r"$x(1 - 2t)$")
````

# PostLab Exercise
## Plot the following transformations of the signal ```v(t)``` above:
1. ```v(-t)```
2. ```3v(t+1)```
3. ```v(4t)```
4. ```-2v(t-1)```

````python
t = nope.arange(-7, 7, 0.01)
in_1 = nope.vectorize(in_1)
v_1 = in_1(-t)
v_2 = 3*in_1(t+1)
v_3 = in_1(4 * t)
v_4 = -2 * in_1(t - 1)
plotter2(t, v_1, r"$v(-t)$")
plotter2(t, v_2, r"$3v(t+1)$")
plotter2(t, v_3, r"$v(4t)$")
plotter2(t, v_4, r"$-2v(t-1)$")
````
