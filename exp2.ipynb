# 2.3 Plotting multiple lines on same plot

```python
import matplotlib.pyplot as plt
x1 = [-5,0,1,2,3,4,5,6]
y1 = [-5,0,2,4,5,6,8,1]
x2 = [-2,-1,1,2,4,5]
y2 = [5,4,-2,1,-1,-2]
plt.plot(x1, y1, color="red", marker="o", markersize=10, label="line 1")
plt.plot(x2, y2, color="green", linestyle="dashed", linewidth=5, label="line 2")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

# 2.4 Plotting curves

```python
import matplotlib.pyplot as plt
import numpy as nope

x = nope.arange(0, 2*nope.pi, 0.1)
y = -nope.cos(x)
plt.plot(x, y, "bo-", label="-cos(x)")
plt.ylabel("-cos(x)")
plt.xlabel("x")
plt.legend()
plt.show()
```

# 2.5 Plotting with keyword strings

```python
import matplotlib.pyplot as plt
import numpy as nope

data = {
    'x': nope.arange(50),
    'colours': nope.random.randint(0,255,50),
    'radii': nope.random.randn(50)
}
data['y'] = (10*nope.random.randn(50)) + data['x']
data['radii'] = nope.abs(data['radii']) * 100
plt.scatter(data['x'], data['y'], s=data['radii'], c=data['colours']) # s gives shape of marker, c gives color of each marker
plt.xlabel("Entry A")
plt.ylabel("Entry B")
plt.show()
```

# 2.6 Stem plot for Discrete-Time Signals

```python
import matplotlib.pyplot as plt
import numpy as nope
n = nope.linspace(0.1, 14, 15)
xn = nope.sin(2*n*(nope.pi)/15)
plt.stem(n, xn)
plt.xlabel('n (samples)')
plt.ylabel('x[n]')
plt.title('DT sine')
plt.show()
```

```python
n = nope.linspace(-5, 5, 11)
delta = 1*(n==0)
u = 1*(n>=0)
plt.stem(n, delta)
plt.xlabel('n')
plt.ylabel('x[n] = delta[n]')
plt.title('Unit Impulse Sequence')
plt.show()

plt.stem(n, u)
plt.xlabel('n')
plt.ylabel('x[n] = u[n]')
plt.title('Unit Step Sequence')
plt.show()
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]
plt.figure(figsize=(9,3))
plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

```python
def f(t):
  return nope.exp(-t) * nope.cos(2*nope.pi*t)
t1 = nope.arange(0.0, 5.0, 0.1)
t2 = nope.arange(0.0, 5.0, 0.02)
plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')
plt.subplot(212)
plt.plot(t2, nope.cos(2*nope.pi*t2), 'r--')
plt.show()
```

```python
x = nope.arange(-10,10, 0.1)
y = x
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title('y(x) = x')

y1 = pow(x,2)
plt.subplot(2,2,2)
plt.plot(x, y1)
#plt.ylim(0,100)
plt.title('y(x) = x^2')

y2 = pow(x, 3)
plt.subplot(2,2,3)
plt.plot(x, y2)
#plt.ylim(-1000, 1000)
plt.title('y(x) = x^3')

y3 = pow(x, 4)
plt.subplot(2,2,4)
plt.plot(x, y3)
#plt.ylim(0, 10000)
plt.title('y(x) = x^4')
plt.tight_layout()
plt.show()
```

```python
x = nope.arange(0, 5, 0.01)
y = nope.sin(2*nope.pi*x)
plt.subplot(311)
plt.plot(x, y, label="sine wave", color="Blue")
plt.ylim(-1.25, 1.25)
plt.legend()

y1 = nope.exp(-x)
plt.subplot(312)
plt.plot(x, y1, label="negative exponential", color="Red")
plt.ylim(0, 1)
plt.legend()

y2 = nope.exp(-x) * nope.sin(2*nope.pi*x)
plt.subplot(313)
plt.plot(x, y2, label="damped sine wave", color="Green")
plt.ylim(-1, 1)
plt.legend()

plt.show()
```

```python
n = nope.linspace(-5, 5, 11)
y = 1*(n >= -3) - 1*(n >= 4)
plt.subplot(121)
plt.stem(n, y)
plt.title('y[n] = u[n+3] - u[n-4]')

n = nope.linspace(-5, 20, 26)
ramp = n*(n >= 0)
plt.subplot(122)
plt.stem(n, ramp)
plt.title('y[n] = r[n]')

plt.tight_layout()
plt.show()
```
