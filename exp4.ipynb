# Problem 1
## Compute DFT of sequence ```x1 = [0, 4, 2, 0]```

````python
import matplotlib.pyplot as plt
import numpy as nope
from scipy.fft import fft

x1 = nope.array([0,4,2,0])
dft_x1 = fft(x1)

plt.subplot(2,1,1)
plt.suptitle(r'DFT of $[0, 4, 2, 0]$')
plt.stem(dft_x1.real)
plt.xlabel(r'$k$')
plt.ylabel(r'$Re\{X[k]\}$')
plt.title("Real Part of X[k]")
plt.show()

plt.subplot(2,1,2)
plt.stem(dft_x1.imag)
plt.xlabel(r'$k$')
plt.ylabel(r'$Im\{X[k]\}$')
plt.title("Imaginary Part of X[k]")
plt.show()

print(f"The DFT of {x1} is {dft_x1}")
````

# Problem 3
## Compute DFT of ```sin(t)```

````python
import matplotlib.pyplot as plt
import numpy as nope
from scipy.fft import fft, fftfreq, fftshift

t = nope.arange(64)
x = nope.sin(t)
sp = fftshift(fft(x))
freq = fftshift(fftfreq(64))

plt.figure(figsize=(4,8))
plt.subplot(2,1,1)
plt.grid()
plt.plot(t,x)
plt.xlabel(r'$t$')
plt.ylabel(r'$sin(t)')

plt.subplot(2,1,2)
plt.grid()
plt.plot(freq, abs(sp))
plt.xlabel(r'$k$')
plt.ylabel(r'$|X[K]|$')
plt.show()
````
